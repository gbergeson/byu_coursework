```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def get_data():
    from sklearn.datasets import fetch_openml
    mnist = fetch_openml('mnist_784', version=1, cache=False)
    return pd.DataFrame(mnist.data / 255.0), pd.Series(mnist.target).astype('int').astype('category')

x,y = get_data()
```

# 1. PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=144)
x_pca = pca.fit_transform(x)
print(x_pca.shape)
```

```python
print(np.cumsum(pca.explained_variance_ratio_)[-1])
```

a. I used 144 components, because it's a perfect square (a lot easier to graph later) and because it still describes 94.5% of the variance in the dataset.

```python
plt.scatter(x_pca[:, 0], x_pca[:, 1],
            c=y, edgecolor='none', alpha=1,
            cmap=plt.cm.get_cmap('plasma', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

b. The trends that I notice are: 
<ol type='a'>
    <li>These digits are pretty clustered together when only visualized in two-dimensional space. This makes it hard to really tell them apart.</li>
    <li>Not only are the clusters really close, they're also overlapping by a lot, complicating the problem even further.</li>
    <li>Even though it's been standardized, there's still a tail off to the right.</li>
</ol>
    
This plot is showing us the best possible projection of our data from 784 dimensional space into 2 dimensional space. The axes of this plot are the two components that explain the most variation of all other components. 

c. You can use PCA to simplify data and reduce its dimensionality to make it easier for other algorithms to handle. You can also use it to compress data, which can be useful for images. It's an unsupervised method, meaning it doesn't reference any labels for its data. Basically, it examines data and looks at the axes or dimensions in the data and calculating how much of the variance in the dataset they each explain. Then, it keeps the top n (n is a hyperparameter) components that explain the most variance. The drawback is that you do lose some of the variance in the dataset, potentially losing some of the trends and patterns in the data.

d. For these plots, I use the raw data and the data we compressed with pca earlier.

# T-SNE

```python
from sklearn.manifold import TSNE
import seaborn as sns

rows = np.arange(70000)
np.random.shuffle(rows)
n_select = 10000

def graph_tsne(perplex, iter, rate):
    tsne = TSNE(n_components=2, perplexity=perplex, n_iter=iter, learning_rate=rate)
    tsne_results = tsne.fit_transform(x_pca[rows[:n_select],:])
    
    df_tsne = pd.DataFrame(tsne_results, columns=['comp1', 'comp2'])
    df_tsne['label'] = y[rows[:n_select]]

    sns.lmplot(x='comp1', y='comp2', data=df_tsne, hue='label', fit_reg=False)
```

```python
graph_tsne(50, 1000, 200)
```

a. This plot is different from the PCA plot in the following ways:
 - Space can be seen between some clusters
 - All the clusters are totally overlaid instead of being only partially overlaid. This makes it seem like it will be even harder to group into clusters
 - The colors aren't as cool. ;)
 
This top graph is the one I'll be comparing all the lower ones to to check robustness for changes in perplexity, number of iterations, and learning rate.

```python
graph_tsne(30, 1000, 200)
```

```python
graph_tsne(5, 1000, 200)
```

b. It's decently robust to changes in perplexity, except when those changes are large. The 50 and 30 graphs look pretty similar, but the 5 graph looks totally different.

```python
graph_tsne(50, 700, 200)
```

```python
graph_tsne(50, 500, 200)
```

c. Again, it's decently robust to changes in number of iterations (probably more robust than it is to changes in perplexity), but you can definitely notice changes when those changes are large. Not much difference between 1000 and 700, but plenty of difference between 1000 and 500. 

```python
graph_tsne(50, 1000, 150)
```

```python
graph_tsne(50, 1000, 100)
```

As for changes in learning rate, it's probably the least robust of the three. Even though this last one looks decently similar to the 200 learning rate one, any change in learning rate can definitely be noticed.

# KMeans

Before we do kmeans, we have to scale the data. (part d)

```python
# Thanks to https://medium.com/@rrfd/standardize-or-normalize-examples-in-python-e3f174b65dfc for this code!

from sklearn.preprocessing import StandardScaler
# Get column names first
names = x.columns
# Create the Scaler object
scaler = StandardScaler()
# Fit your data on the scaler object
scaled_x = scaler.fit_transform(x)
scaled_x = pd.DataFrame(scaled_x, columns=names)
```

```python
from sklearn.cluster import KMeans

sse = []
list_k = list(range(7, 13))

for k in list_k:
    km = KMeans(n_clusters=k)
    km.fit(scaled_x)
    sse.append(km.inertia_)

# Plot sse against k
plt.figure(figsize=(6, 6))
plt.plot(list_k, sse, '-o')
plt.xlabel(r'Number of clusters *k*')
plt.ylabel('Sum of squared distance');
```

a, b. I centered my choices for k around 10 because there are 10 digits in the data, making it logical that there will be 10 clusters of data. I evaluated my clustering by using the "elbow method." It doesn't really look like 10 is a super good choice from the elbow method, but it definitely makes the most sense when approaching the problem from the outside. What this means, probably, is that the data is pretty similar throughout, making it difficult to really cluster well.

c. Visualization of clusters using t-SNE:

```python
km = KMeans(n_clusters=10)
km.fit(scaled_x)
```

```python
from sklearn.manifold import TSNE
import seaborn as sns

rows = np.arange(70000)
np.random.shuffle(rows)
n_select = 10000

tsne = TSNE(n_components=2, perplexity=40, n_iter=1000, learning_rate=200)
tsne_results = tsne.fit_transform(scaler.fit_transform(x)[rows[:n_select],:])
    
df_tsne = pd.DataFrame(tsne_results, columns=['comp1', 'comp2'])
df_tsne['label'] = km.labels_[rows[:n_select]]

sns.lmplot(x='comp1', y='comp2', data=df_tsne, hue='label', fit_reg=False)
```

d. Yes.

e. Pretty robust! The clusters are fairly distinguishable from each other, but they do have some overlap. Fairly understandable, since we're only looking at two dimensions.
